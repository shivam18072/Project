```python
data['c_id'] = data.CustomerID.astype('object')
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format='%m/%d/%Y %H:%M')
```

```python
data['day'] = data['InvoiceDate'].dt.day
```

```python
data.head()
```

```python
data['invoice_time'] = data['InvoiceDate'].map(lambda x : 100 * x.year + x.month)
```

```python
data.head()
```

```python
data['invoice_time'] = data.invoice_time.astype('object')
```

```python
data.info()
```

```python
data.describe(include = 'all')
```

```python
data = data[data.UnitPrice > 0]
```

```python
data.describe(include = 'all')
```

```python
# how many orders made by customer ?

temp1 = data.groupby(by=['c_id'], as_index = False)['InvoiceNo'].count()
```

```python
temp1.sort_values(by = ['InvoiceNo'],ascending = False)
```

```python
temp1[temp1['InvoiceNo'] == 1]
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.subplots(figsize = (15,6))
plt.xlabel('Customer ID')
plt.ylabel('Number of orders')
plt.plot(temp1.c_id,temp1.InvoiceNo)
plt.show()
```

```python
# How much money spent by the customer 

data['amount_spent'] = data['Quantity'] * data['UnitPrice']
```
